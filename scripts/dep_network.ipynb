```python
# FEATURES AND DEPENDENCIES
dependencies = {
    "Season": [],
    "Precipitation": [],
    "Mean property value": [],
    "Building age": [],
    "Holiday flag": [],
    "Emergency response time": ["Precipitation", "Population density", "Main road distance", "Flood depth"],
    "Ambulance handover time": ["Season"],
    "Hospital bed availability": ["Season"],
    "English proficiency": [],
    "Vehicle rate": [],
    "Water density": [],
    "Watercourse distance": [],
    "Elevation": [],
    "Impervious surface area": [],
    "Historic flood flag": [],
    "Transport": [],
    "Main road distance": [],
    "Hospital distance": [],
    "Buildings density": [],
    "Population density": [],
    "Emergency services": [],
    "Deprivation rank": [],
    "Education": [],
    "Residential": [],
    "Commercial": [],
    "Industrial": [],
    "Agriculture": [],
    "Infrastructure": [],
    "Soil moisture": ["Season", "Precipitation", "Flood depth"],
    "Flood depth": ["Precipitation", "Impervious surface area", "Soil moisture", "Watercourse distance"],
    "Damage cost": ["Flood depth", "Transport", "Residential", "Commercial", "Industrial", "Agriculture", "Infrastructure"],
    "Age": [],
    "Highest level of qualification": [],
    "Accomodation type": [],
    "Household size": [],
    "Economic activity status": ["Age"],
    "NS-SeC": ["Age", "Economic activity status"],
    "Mean income": [],
    "Low income flag": ["Mean income", "NS-SeC"],
    "Adults employed in household": ["Household size"],
    "Disabled people in household": ["Household size"],
    "People with long-term health in household": ["Household size"],
    "Deprived in education": ["Highest level of qualification"],
    "Deprived in employment": ["Adults employed in household", "NS-SeC"],
    "Deprived in health and disability": ["People with long-term health in household", "Disabled people in household"],
    "People per room in household": ["Household size"],
    "Occupancy rating for rooms": ["People per room in household"],
    "Deprived in housing": ["People per room in household", "Occupancy rating for rooms"],
    "Household deprivation": ["Deprived in education", "Deprived in employment", "Deprived in health and disability", "Deprived in housing"],
    "Tenure of household": [],
    "Internet flag": ["Age", "Household size"],
    "Home insurance flag": [],
    "Health insurance flag": [],
    "Household": ["Tenure of household", "Accomodation type", "Household size", "Internet flag", "Household deprivation", "Low income flag", "Home insurance flag", "Health insurance flag"],
    "Home insurance rate": [],
    "Mean income of households": ["Mean income"],
    "Disability rate": [],
    "General health rate": [],
    "Elderly rate": ["Age"],
    "Children rate": ["Age"],
    "Physical vulnerability": ["Elevation", "Impervious surface area", "Water density", "Watercourse distance", "Transport", "Building age", "Buildings density"],
    "Socioeconomic vulnerability": ["Elderly rate", "Disability rate", "Children rate", "General health rate", "English proficiency", "Household", "Mean property value", "Deprivation rank", "Education"],
    "Preparedness": ["Emergency response time", "Ambulance handover time", "Hospital bed availability", "Vehicle rate", "Hospital distance", "Emergency services", "Warning Issued"],
    "Recovery capacity": ["Mean income of households", "Home insurance rate"],
    "Exposure": ["Historic flood flag", "Holiday flag", "Population density"],
    "Warning Issued": [],
    "Impact score": ["Exposure", "Physical vulnerability", "Damage cost", "Precipitation", "Socioeconomic vulnerability", "Preparedness", "Recovery capacity"],
}
```

```python
# DIRECTED GRAPH
G = nx.DiGraph()
G.add_nodes_from(dependencies.keys())
for node, deps in dependencies.items():
    for dep in deps:
        G.add_edge(dep, node)
```

```python
# COLOUR CODING

cmap = plt.cm.Set3
color_map = []
default_color = "lightgrey"
for node in G.nodes():
    #clean_node = remove_num(node)
    if node == "Impact score":
        color_map.append(cmap(0))
    elif node in dependencies["Impact score"]:
        color_map.append(cmap(1))
    elif node in dependencies["Physical vulnerability"]:
        color_map.append(cmap(2))
    elif node in dependencies["Socioeconomic vulnerability"]:
        color_map.append(cmap(3)) 
    elif node in dependencies["Preparedness"]:
        color_map.append(cmap(4)) 
    elif node in dependencies["Recovery capacity"]:
        color_map.append(cmap(5)) 
    elif node in dependencies["Exposure"]:
        color_map.append(cmap(6)) 
    elif node in dependencies["Household"]:
        color_map.append(cmap(7))
    else:
        color_map.append(default_color)
```

```python

# Draw the graph

'''shells = [
    ["60. Impact score"],
    dependencies['60. Impact score'],
    [n for n in G.nodes() if n != "60. Impact score" and n not in dependencies['60. Impact score']],
]
pos = nx.shell_layout(G, nlist=shells)'''

#impact_dep = [name_to_node[x] for x in dependencies["Impact score"]]
impact_dep = [x for x in dependencies["Impact score"]]

rest = [n for n in G.nodes() if n not in ["Impact score"]+impact_dep]
shell_pos = nx.shell_layout(G, nlist=[["Impact score"], impact_dep])

pos = nx.spring_layout(
    G,
    pos=shell_pos,     # initial positions include the fixed shell positions
    k=0.25,
    fixed=["Impact score"]+impact_dep  # these nodes will not move
)

'''impact_dep = [name_to_node[x] for x in dependencies["60. Impact score"]]
rest = [n for n in G.nodes() if n not in ["60. Impact score"]+impact_dep]
shells = [
    ["60. Impact score"],
    impact_dep,
    rest
]

pos = nx.shell_layout(G, nlist=shells, scale=2.0)'''



# DRAW
plt.figure(figsize=(16, 16))
nx.draw_networkx_nodes(G, pos, node_color=color_map, node_size=2600, alpha=0.80)
nx.draw_networkx_labels(G, pos, font_size=8, font_weight='bold')
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=12, edge_color="k", width=2.5,min_source_margin=25,min_target_margin=25)

plt.title("Feature Dependency Network", fontsize=14, fontweight="bold")
plt.axis("off")
plt.tight_layout()
plt.savefig("dependency_network.png", dpi=300)
plt.show()
```
